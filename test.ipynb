```python
import re
import pandas as pd
```

```python
f = open('Chat1.txt','r',encoding='utf-8')
```

```python
data = f.read()
data
```

```python
pt1 = '\d{1,2}/\d{1,2}/\d{1,2},\s\d{1,2}:\d{2}\s-\s'
pt2 = '\d{1,2}/\d{1,2}/\d{2},\s\d{1,2}:\d{2}(?:\s[ap]m)?\s-\s'
```

```python
def find_pt(data):
    p1 = re.findall(pt1,data)
    p2 = re.findall(pt2,data)

    if len(p1) != 0:
        return pt1
    else:
        return pt2
```

```python
pattern = find_pt(data)
```

```python
masseges = re.split(pattern , data)[1:]
masseges
```

```python
dates = re.findall(pattern,data)
dates
```

```python
df = pd.DataFrame({'user_message':masseges , 'message_date':dates})

df['message_date'] = pd.to_datetime(df['message_date'],format='%d/%m/%y, %I:%M %p - ')
df.rename(columns={'message_date':'date'},inplace=True)
```

```python
df.shape
```

```python
# Separate users and message

users =[]
messsages = []

for mssg in df['user_message']:
    entry = re.split('([\w\W]+?):\s' , mssg)
    if entry[1:]:
        users.append(entry[1])
        messsages.append(entry[2])
    else:
        users.append('group_notification')
        messsages.append(entry[0])

df['users'] = users
df['message'] = messsages

df.drop(columns=['user_message'],inplace=True)
```
